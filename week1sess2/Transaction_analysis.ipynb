```python
def transaction(transactlist):
  if transactlist:
    total = sum(transactlist)
    average = sum(transactlist) / len(transactlist)
    num = len(transactlist)

    print(f'sum : {total}')
    print(f'average : {average}')
    print(f'num of transactions : {num}')
  else:
    print('empty list')

transaction([140,280,399,420,560,600, 850])


```
