#DATA CLEANING

```python
# Missing values per column
data_clean.isna().sum()
```

###HANDLING MISSING VALUES

```python
# FILLING THE MISSING ROW UNDER ITEM TYPE WITH WINE
data_clean['ITEM TYPE'] = data_clean['ITEM TYPE'].fillna('WINE')
```

```python
data_clean['RETAIL SALES'] = data_clean['RETAIL SALES'].fillna(data_clean['RETAIL SALES'].mode()[0])
```

```python
# FILL MISSING VALUES UNDER THE COLUMN 'SUPPLIER'
data_clean['SUPPLIER'] = data_clean['SUPPLIER'].fillna('UNKNOWN')
# CHECK IF THERE ARE STILL MISSING VALUES
data_clean.isna().sum()
```

```python
data_clean.info()
```

### CHECK FOR DUPLICATES

```python
# CHECK FOR DUPLICATES
data_clean.duplicated().sum()
```

###FEATURE ENGINEERING

```python
# CREATE A NEW COLUMN 'DATE' AND MAKE IT A DATETIME DATATYPE
data_clean['DATE'] = data_clean['YEAR'].astype(str) + '-' + data_clean['MONTH'].astype(str)
data_clean['DATE'] = pd.to_datetime(data_clean['DATE'])
data_clean.info()
```

```python
# CREATE TOTAL SALES COLUMN TO SEE TOTAL SALES FROM EACH TRANSACTIONS
data_clean['TOTAL SALES'] = data_clean['RETAIL SALES'] + data_clean['WAREHOUSE SALES']
data_clean.head()
```

```python
# SAVE TO A CSV FILE
data_clean.to_csv('Cleaned_Sales_Data.csv', index=False)
```
